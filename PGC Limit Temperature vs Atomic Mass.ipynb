```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
h = 6.62607015e-34  # Planck's constant in J*s
kB = 1.380649e-23  # Boltzmann constant in J/K
lambda_laser = 780e-9  # Wavelength of the laser in meters (780 nm)

# Define a range of atomic masses, from Lithium to Cesium (in atomic mass units converted to kg)
atomic_masses_amu = np.linspace(6, 133, 100)  # From Lithium to Cesium
atomic_masses_kg = atomic_masses_amu * 1.66053906660e-27  # Convert amu to kg

# Recalculate the recoil limit temperatures for different atomic masses
T_p_masses = h**2 / (kB * atomic_masses_kg * lambda_laser**2)

# Plotting without the title
plt.figure(figsize=(10, 6))
plt.plot(atomic_masses_amu, T_p_masses, label='Polarization Gradient Cooling Limit')
plt.xlabel('Atomic Mass (amu)')
plt.ylabel('Cooling Limit Temperature (K)')
plt.grid(True)
plt.yscale('log')  # Log scale to better visualize changes
plt.legend()
plt.show()
```
