```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
lambda_wave = 780e-9  # Wavelength of the laser for Rubidium cooling (780 nm)

# Spatial domain for the simulation
z = np.linspace(0, 2 * lambda_wave, 500)  # Two wavelengths long

# Calculate the electric fields of two counter-propagating waves
E0 = 1  # Amplitude of electric fields
k = 2 * np.pi / lambda_wave  # Wave number

# Electric fields of the waves traveling in +z and -z directions
E_plus = E0 * np.exp(1j * k * z)
E_minus = E0 * np.exp(-1j * k * z)

# Standing wave formation
E_total = E_plus + E_minus

# Intensity of the standing wave
intensity = np.abs(E_total)**2

# Polarization varies along z, let's assume it cycles through linear to circular
# Linear polarization at 0, circular at lambda/4, and back to linear at lambda/2
# This repeats over the wavelength of the standing wave
polarization_state = np.cos(2 * np.pi * z / lambda_wave)

# Plotting
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Position (meters)')
ax1.set_ylabel('Intensity (arbitrary units)', color=color)
ax1.plot(z, intensity, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
color = 'tab:blue'
ax2.set_ylabel('Polarization State', color=color)  # we already handled the x-label with ax1
ax2.plot(z, polarization_state, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.title('Standing Wave Pattern and Polarization State Along z-axis')
plt.show()
```
